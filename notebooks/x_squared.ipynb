```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units=4, input_shape=[1], activation='tanh'), tf.keras.layers.Dense(units=1)])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
xs = np.array([-4.0, -2.0, -1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([16.0, 4.0, 1.0, 0.0, 1.0, 4.0, 9.0, 16.0], dtype=float)
```

```python
model.fit(xs, ys, epochs=1000)
```

```python
model.predict([5.0])
```
